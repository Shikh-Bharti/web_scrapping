```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Web Scraping
def scrape_amazon_data(query, num_pages=5):
    base_url = "https://www.amazon.in/s?k="
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/85.0.4183.121 Safari/537.36"
    }
    product_list = []

    for page in range(1, num_pages + 1):
        url = f"{base_url}{query}&page={page}"
        response = requests.get(url, headers=headers)
        soup = BeautifulSoup(response.text, 'html.parser')

        products = soup.find_all("div", {"data-component-type": "s-search-result"})

        for product in products:
            try:
                name = product.h2.text.strip()
                price = product.find("span", "a-price-whole")
                price = int(price.text.replace(',', '')) if price else None
                rating = product.find("span", {"class": "a-icon-alt"})
                rating = float(rating.text.split()[0]) if rating else None
                reviews = product.find("span", {"class": "a-size-base"})
                reviews = int(reviews.text.replace(',', '')) if reviews else None

                product_list.append({
                    "Name": name,
                    "Price": price,
                    "Rating": rating,
                    "Reviews": reviews
                })
            except Exception as e:
                continue

    return pd.DataFrame(product_list)

# Step 2: Data Cleaning
def clean_data(df):
    # Drop rows with missing values
    df.dropna(inplace=True)

    # Remove outliers in price and reviews
    df = df[(df['Price'] < df['Price'].quantile(0.95)) & (df['Price'] > df['Price'].quantile(0.05))]
    df = df[(df['Reviews'] < df['Reviews'].quantile(0.95)) & (df['Reviews'] > df['Reviews'].quantile(0.05))]

    return df

# Step 3: Data Analysis & Visualization
def visualize_data(df):
    # Rating vs Price
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x="Price", y="Rating", size="Reviews", hue="Reviews", data=df, palette="viridis")
    plt.title("Rating vs Price (Bubble Size = Reviews)")
    plt.xlabel("Price (INR)")
    plt.ylabel("Rating")
    plt.show()

    # Distribution of Prices
    plt.figure(figsize=(10, 6))
    sns.histplot(df['Price'], kde=True, bins=30, color="blue")
    plt.title("Distribution of Product Prices")
    plt.xlabel("Price (INR)")
    plt.ylabel("Frequency")
    plt.show()

    # Top 10 Most Reviewed Products
    top_products = df.nlargest(10, 'Reviews')
    plt.figure(figsize=(12, 6))
    sns.barplot(x="Reviews", y="Name", data=top_products, palette="mako")
    plt.title("Top 10 Most Reviewed Products")
    plt.xlabel("Number of Reviews")
    plt.ylabel("Product Name")
    plt.show()

# Main Execution
if __name__ == "__main__":
    query = "laptop"  # Change query to desired product
    raw_data = scrape_amazon_data(query, num_pages=3)
    print("Raw Data Scraped:", raw_data.head())

    clean_data = clean_data(raw_data)
    print("Cleaned Data:", clean_data.head())

    visualize_data(clean_data)
```
